# Nueva sección

```python
df['sentimiento'].value_counts()
```

```python
def contar_palabras(texto):
  l = texto.split(' ')
  return len(l)
```

```python
df['word_count'] = df['review_es'].apply(contar_palabras)
df.head()
```

```python
df = df.rename(columns={'review_es':'text'})
```

```python
df['word_count'].max()
```

```python
df['word_count'].min()
```

```python
df
```

```python
df2=df['sentimiento'].replace('positivo',1).replace('negativo',0)
df['labels'] = df2
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['word_count'], bins=10)
plt.show()
```

```python
df3 = df[['text', 'labels']]
df3
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df3, test_size=0.3)

print(train.head())
print(test.head())
```

```python
train.to_csv('/content/gdrive/MyDrive/Hanagon/train.csv', index=False)
test.to_csv('/content/gdrive/MyDrive/Hanagon/test.csv', index=False)
```
